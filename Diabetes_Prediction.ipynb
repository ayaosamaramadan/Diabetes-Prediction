Much Better

```python
def check_and_handle_outliers(column,df):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)

  IQR= Q3 - Q1

  lower_limit = Q1 - 1.5 * IQR
  upper_limit = Q3 + 1.5 * IQR

  for rows in range(len(df)):
    if df.loc[rows,column] > upper_limit:
      df.loc[rows,column] = upper_limit

    elif df.loc[rows,column] < lower_limit:
      df.loc[rows,column] = lower_limit

```

```python
for column in df.columns:

  if df[column].dtypes == 'float64' or df[column].dtypes == 'int64':
    if column == 'diabetes':
      print("diabetes")
    else:
      check_and_Manipulate_Negativity(column, df)
      check_and_handle_outliers(column,df)
      print("Negativity and Outliers had gone !")
```

**Preprocessing**

```python
X = df.drop(columns='diabetes',axis=1)
y = df['diabetes']
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=33)
```

```python
from sklearn.metrics import accuracy_score
```

**ٌRandom Forest**

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier()
RF.fit(X_train,y_train)
```

```python
y_pred=RF.predict(X_test)
Score1=accuracy_score(y_test,y_pred)
print("Test score : ",Score1*100,"%")
```
